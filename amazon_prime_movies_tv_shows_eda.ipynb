## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Remove duplicate values


titles = titles.drop_duplicates()
credits = credits.drop_duplicates()


# 2. Check missing values again
print("Missing values in Titles dataset:")
print(titles.isnull().sum())

print("\nMissing values in Credits dataset:")
print(credits.isnull().sum())



# 3. Handle missing values

# Fill missing description with 'Unknown'
titles['description'] = titles['description'].fillna('Unknown')
# Fill missing age certification with 'Not Rated'
titles['age_certification'] = titles['age_certification'].fillna('Not Rated')

# Fill missing imdb_score with median
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].median())


# Fill missing tmdb_score with median
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].median())



# 4. Drop columns with too many missing values
titles = titles.drop(columns=['seasons'], errors='ignore')
# 8514 out of 9871 valus missing in this column


# 5. Convert data types if needed
titles['release_year'] = titles['release_year'].astype(int)
# release_year showed object as data type
# 6. Final dataset shape
print("\nFinal shape of Titles dataset:", titles.shape)
print("Final shape of Credits dataset:", credits.shape)



# 7. Preview cleaned dataset
titles.head()
````

````python
# Save cleaned datasets

titles.to_csv('/content/cleaned_titles.csv', index=False)
credits.to_csv('/content/cleaned_credits.csv', index=False)
````
